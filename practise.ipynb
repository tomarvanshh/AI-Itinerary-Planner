## Reading CSV with io.StringIO and csv.DictReader

The proper way to read CSV content:
1. Read file as string
2. Create StringIO object (treats string as file-like object)
3. Reset position to start (seek(0))
4. Use DictReader to parse CSV into dictionaries

```python
# Method 1: Basic approach - Read file and create StringIO
with open('uttarakhand_tourism_dataset.csv', 'r') as file:
    csv_content = file.read()

# Create StringIO object from the CSV content
mydata = io.StringIO(csv_content)

# IMPORTANT: Reset position to start (needed after operations on StringIO)
mydata.seek(0)

# Create DictReader - treats each row as an OrderedDict
dict_reader = csv.DictReader(mydata)

# Print all rows
print("=== All Rows ===")
for row in dict_reader:
    print(row)
```

```python
# Method 2: Better approach - Reset and print specific fields
mydata.seek(0)  # Reset to beginning
dict_reader = csv.DictReader(mydata)

print("\n=== Printing Specific Columns ===")
for row in dict_reader:
    # Access specific columns by key
    print(f"Name: {row.get('name', 'N/A')}, Location: {row.get('location', 'N/A')}")
```
